```python
image_path = r"/home/tdjg/Documents/DeepSolaris/Images/Tiles/Heerlen75x75"
image_paths = [os.path.join(image_path, filename) for filename in os.listdir(image_path)]
```

```python
rgb_year_images = [image_path for image_path in sorted(image_paths) if "rgb" in image_path and "mask" not in image_path]
rgb_year_images[0:10]
```

```python
rgb_2016 = [image_path for image_path in rgb_year_images  if image_path.endswith("2016.tiff")]
rgb_2016
```

```python
rgb_2017 = [image_path for image_path in rgb_year_images  if image_path.endswith("2017.tiff")]
rgb_2017
```

```python
len(rgb_2016), len(rgb_2017)
```

```python
images_2016 = [cv.imread(filename) for filename in rgb_2016]
```

```python
images_2017 = [cv.imread(filename) for filename in rgb_2017]
```

```python
images_2016[10].shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def normalized_hist(hist_image, channel, mask=None):
    hist = cv.calcHist([hist_image], [channel], mask, [256], [0, 255])  
    number_of_pixels = hist_image.size if mask is None else mask[mask > 0].size
    return hist / number_of_pixels

def all_bands_hist(hist_image, mask=None):
    all_bands_hist = []
    for i in range(3):
        all_bands_hist.extend(normalized_hist(hist_image, i, mask))
    return np.array(all_bands_hist)
```

```python
hists_2016 = np.array([[normalized_hist(image, channel) for image in images_2016] for channel in range(3)])

hists_2016 = hists_2016.reshape(hists_2016.shape[0:3])
hists_2016.shape
```

```python
summed_hist_2016 = hists_2016.sum(axis=1)
summed_hist_2016.shape
```

```python
for i, c in enumerate(["b", "g", "r"]):
    plt.plot(summed_hist_2016[i, :], color=c)
```

```python
hists_2017 = np.array([[normalized_hist(image, channel) for image in images_2017] for channel in range(3)])
hists_2017 = hists_2017.reshape(hists_2016.shape[0:3])
hists_2017.shape
```

```python
summed_hist_2017 = hists_2017.sum(axis=1)
for i, c in enumerate(["b", "g", "r"]):
    plt.plot(summed_hist_2017[i], color=c)
```

```python
_, ax = plt.subplots(3, 1, figsize=(10,10))

for i,c in enumerate(["blue", "green", "red"]):
    ax[i].set_title("{} 2016 vs. {} 2017".format(c,c))
    ax[i].plot(summed_hist_2016[i], label="2016")
    ax[i].plot(summed_hist_2017[i], color="r", label="2017")
    ax[i].legend()
```
