```python
import os
import numpy as np
import cv2 as cv
```

```python
image_path = r"/home/tdjg/Documents/DeepSolaris/Images/Tiles/Heerlen75x75"
image_paths = [os.path.join(image_path, filename) for filename in os.listdir(image_path)]
```

```python
len(image_paths)
```

```python
image_paths[0:10]
```

```python
rgb_year_images = [image_path for image_path in image_paths if "rgb" in image_path and "mask" not in image_path]
rgb_year_images[0:10]
```

```python
rgb_2016 = [image_path for image_path in rgb_year_images  if image_path.endswith("2016.tiff")]
rgb_2016
```

```python
rgb_2017 = [image_path for image_path in rgb_year_images  if image_path.endswith("2017.tiff")]
rgb_2017
```

```python
images_2016 = [cv.imread(filename) for filename in rgb_2016]
images_2017 = [cv.imread(filename) for filename in rgb_2017]
```

```python
images_2016[10].shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
hists_2016 = [cv.calcHist(images_2016, [i], None, [256], [0,255]) for i in range(3)]

for i, c in enumerate(["b", "g", "r"]):
    plt.plot(hists_2016[i], color=c)
```

```python
hists_2017 = [cv.calcHist(images_2017, [i], None, [256], [0,255]) for i in range(3)]

for i, c in enumerate(["b", "g", "r"]):
    plt.plot(hists_2017[i], color=c)
```

```python
_, ax = plt.subplots(3, 1, figsize=(10,10))

for i,c in enumerate(["blue", "green", "red"]):
    ax[i].set_title("{} 2016 vs. {} 2017".format(c,c))
    ax[i].plot(hists_2016[i])
    ax[i].plot(hists_2017[i], color="r")
```
